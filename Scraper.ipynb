```python
import urllib
import shutil
import requests
import re
from StringIO import StringIO
from PIL import Image
import profile
```

```python
# download all coronal images
url = "http://www.neuroanatomy.ca/MRIs/mri_coronal.html?id=1"
f = urllib.urlopen(url)
gg = f.read()
ix = [m.start() for m in re.finditer('s/(.*).jpg',gg)]
for i in ix:
    k = 1
    j = gg[i:i+20]
    j = re.split('/',j)[1]
    j = re.split('g',j)[0]
    j = j + 'g'
    url = 'http://www.neuroanatomy.ca/MRIs/mri_coronal_files/'+j
    r = requests.get(url)
    i = Image.open(StringIO(r.content))
    i.save(j)
```

```python
# download slices
url = "http://www.neuroanatomy.ca/cross_sections/sections_coronal.html"
f = urllib.urlopen(url)
gg = f.read()
ix = [m.start() for m in re.finditer('files/(.*)g',gg)]
for i in ix:
    k = 1
    j = gg[i:i+50]
    j1 = re.split('/',j)[1]
    if re.search('labels',j):
        j2 = re.split('/',j)[2]
        j = j1+'/'+j2
        j = re.split('\'',j)[0]
        fname = re.split('/',j)[1]
    else:
        j = re.split('"',j1)[0]
        j = re.split('\'',j)[0]
        fname = j
    url = 'http://www.neuroanatomy.ca/cross_sections/sections_coronal_files/'+j
    r = requests.get(url)
    i = Image.open(StringIO(r.content))
    i.save(fname)
```
